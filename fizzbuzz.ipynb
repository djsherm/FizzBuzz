```python
for i in range(1,101):
    output = ''
    speak = (3, 5, 13, 15)
    shout = ('Fizz', 'Buzz', 'Biff', 'Boom')
    noSpeak = (20, 30)
    
    for j in range(len(speak)): #checks if you should shout any of the noises
        if (i % speak[j] == 0):
            output += shout[j]
    
    for k in range(len(noSpeak)):
        if (i % noSpeak[k] == 0):
            output = str(i) # override previous shouting with just saying the number
    
    if output == '': #simply say the number if all else fails
        output = str(i)
    
    print(output)
```
